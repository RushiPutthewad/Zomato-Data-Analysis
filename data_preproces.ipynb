```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Remove text values and convert to numeric
df['RATING_CLEAN'] = pd.to_numeric(df['RATING'], errors='coerce')

# Remove rows with non-numeric ratings
df_clean = df.dropna(subset=['RATING_CLEAN'])

# Check distribution
print(f"Mean: {df_clean['RATING_CLEAN'].mean():.2f}")
print(f"Median: {df_clean['RATING_CLEAN'].median():.2f}")
print(f"Skewness: {stats.skew(df_clean['RATING_CLEAN']):.2f}")

# Plot histogram
plt.hist(df_clean['RATING_CLEAN'], bins=30, alpha=0.7)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Determine if normal or skewed
skewness = stats.skew(df_clean['RATING_CLEAN'])
if abs(skewness) < 0.5:
    print("Distribution is approximately normal")
elif skewness > 0.5:
    print("Distribution is positively skewed (right-skewed)")
else:
    print("Distribution is negatively skewed (left-skewed)")
```

```python
# Drop single column
df = df.drop('RATING_CLEAN', axis=1)
```

```python
# Convert RATING to numeric, text values become NaN
df['RATING'] = pd.to_numeric(df['RATING'], errors='coerce')

# Calculate mean of numeric values only
mean_rating = df['RATING'].mean()

# Fill NaN (text values) with mean
df['RATING'] = df['RATING'].fillna(mean_rating)

print(f"Mean rating used for replacement: {mean_rating:.2f}")
print(f"Unique values after replacement: {sorted(df['RATING'].unique())}")
```

```python
# Calculate duplicate percentage
original_count = len(df)
df_clean = df.drop_duplicates()
clean_count = len(df_clean)
duplicates_removed = original_count - clean_count
duplicate_percentage = (duplicates_removed / original_count) * 100

print(f"Original rows: {original_count}")
print(f"After removing duplicates: {clean_count}")
print(f"Duplicates removed: {duplicates_removed}")
print(f"Percentage of duplicates removed: {duplicate_percentage:.2f}%")
```
